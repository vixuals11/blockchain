```python
#function for crossover of two parents
def CrossoverofTwoParent(parent1,parent2):
#     print("P1",parent1)
#     print("P2",parent2)
    crossover_point=random.randint(2,int(len(parent1)/2))
#     print("\n\n\n   Point of Cross :::::    ",pointofcross)
#     for i in range(pointofcross):
#         print(parent1[i],parent2[i])=print(parent2[i],parent1[i])
       
    newcomb1 = []
    for i in range(0, crossover_point):
        newcomb1.append(parent1[i])
    for i in range(crossover_point, len(parent1)):
        newcomb1.append(parent2[i])

    newcomb2 = []
    for i in range(0, crossover_point):
        newcomb2.append(parent2[i])
    for i in range(crossover_point, len(parent1)):
        newcomb2.append(parent1[i])
    
#     print("C1",newcomb1)
#     print("C2",newcomb2)
    newlist=[]
    newlist.append(newcomb1)
    newlist.append(newcomb2)
#     if parent1.all not in population:
#         newlist.append(parent1)
#     if parent1.all not in population:
#         newlist.append(parent2)

    return newlist
#     print("Printing New List  \n\n",newlist,"\n\n")
#function for cross over of a generation
def genCross(population):
    print("Crossing over the generation to create new generation...")
    size=len(population)
#     print(size)
    i=0
    while i in range(0,size):
        newlist=[]
        
        newlist=CrossoverofTwoParent(population[i],population[i+1])
        
        population.extend(newlist)
        i=i+2
genCross(population)
```

```python
from numpy.ma.core import empty
def Check_Hard_Constraints(population):
    
    HardFitChromosomes = []
    goodones = []
    AllFitScore = [] #Fitness score for all chormosoes
    currScore = 0 # fitness score for current chromosome
    
    while(not len(HardFitChromosomes)):
      for x in population:
          
          if(NoTeacherSameTime(x)):
              currScore += 1
          # print("Current Score",currScore)
          if(NoSectionSameTime(x)):
              currScore += 1
              
          if(NoClassroomSametime(x)):
              currScore += 1
  #         print("Current Score",currScore)
          if(consecutive(x)):
              currScore += 1
              
          if(NoClassBeforeAfter(x)):
              currScore += 1
  #         print("Current Score",currScore)
          if(weekend(x)):
              currScore += 1
  #         print("Current Score",currScore)
          if(currScore >=4):
            goodones.append(x)
          if(currScore == 6):
              HardFitChromosomes.append(x)
              print("Current Score",currScore)
          # else:
              # print("Not passed")
              
          
          currScore = 0
      population.extend(goodones)
      genCross(population)
      Mutation(population)
    return HardFitChromosomes

Check_Hard_Constraints(population)
```
